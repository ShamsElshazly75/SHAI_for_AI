```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Sleep_health_and_lifestyle_dataset.csv')
df.head()
```

```python
df.info()
```

### The data is clean

```python
df.describe()
```

```python
df['Occupation'].value_counts()
```

### We will replace the long job title with another title

```python
df["Occupation"] = df["Occupation"].replace({'Sales Representative': 'Salesperson'})

df["Occupation"] = df["Occupation"].replace({'Software Engineer': 'Engineer'})
```

```python
sns.heatmap(df.corr())
```

# ERD

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# Barplot for average sleep duration by occupation
sns.barplot(x='Occupation', y='Sleep Duration', data=df, ax=axes[0])
axes[0].set_title('Average Sleep Duration by Occupation')
axes[0].set_xlabel('Occupation')
axes[0].set_ylabel('Average Sleep Duration (hours)')

# Barplot for average quality of sleep by occupation
sns.barplot(x='Occupation', y='Quality of Sleep', data=df, ax=axes[1])
axes[1].set_title('Average Quality of Sleep by Occupation')
axes[1].set_xlabel('Occupation')
axes[1].set_ylabel('Average Quality of Sleep')

plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x='Gender', y='Age', data=df, palette='pastel')
plt.title('Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

# Boxplot for Stress Level by Gender
sns.boxplot(x='Gender', y='Stress Level', data=df, palette='pastel', ax=axes[0])
axes[0].set_title('Stress Level by Gender')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Stress Level')

# Boxplot for Stress Level by Gender
sns.boxplot(x='Gender', y='Stress Level', data=df, palette='pastel', ax=axes[1])
axes[1].set_title('Stress Level by Gender')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Stress Level')

plt.tight_layout()
plt.show()
```

## We note that Men have higher levels of stress than women

```python
sns.countplot(data= df, x= 'Stress Level',hue ='BMI Category')
plt.title('Relationship Between BMI_Category and Stress Level')
plt.show()
```

## Stress increases with those who are overweight

```python
plt.figure(figsize= (15,10))
sns.countplot(data= df, x= 'Occupation',hue = 'Sleep Disorder')
plt.title('Relationship Between Occupation and Sleep_Disorder')
plt.show()
```

## The incidence of insomnia increases among nurses, and the incidence of apnea increases among teachers and salesperson

```python
sns.displot(data=df, x='Physical Activity Level', hue="Stress Level", multiple="stack")
```

## There is no strong relationship between physical acticivty level and stress level

```python
sns.displot(data=df, x="Quality of Sleep", multiple="stack", kind="kde")
```

```python
g = sns.relplot(x=df['Quality of Sleep'], y=df['Quality of Sleep'].index, col=df['Gender'])
g.set_axis_labels("Quality of Sleep", "Count")
```
